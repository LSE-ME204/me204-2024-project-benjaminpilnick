# Getting genres of all artist in the certain playlist 

```python


#functin that gets the genres of the entire playlist, given a certain uri of the playlist.
def get_genres(uri):

    #this gets the playlist 
    results = spotify.playlist(uri)

    #getting artist_id's
    artist_id = []
    for tracks in range(len(results['tracks']['items'])): 
        artist_id.append(results['tracks']['items'][tracks]['track']['artists'][0]['uri'])


    #getting genre's of each id. 

    
    genres = []
    for artist in artist_id: 
       genres.append(spotify.artist((artist))['genres'])

    return genres
```

```python

#function that gets the categories of artists in a playlist
def get_genres(uri):
    # Retrieves the playlist and extracts artist URIs
    results = spotify.playlist(uri)
    artist_ids = [track['track']['artists'][0]['uri'] for track in results['tracks']['items']]
    
    # Retrieves and returns genres for each artist
    return [spotify.artist(artist)['genres'] for artist in artist_ids]
```
